```python
from tkinter import*
import string
import random
from tkinter import messagebox

root=Tk()
root.geometry('400x405')
root.resizable(1,1)
root.title('Password Generator')

def generate_password():
    username = x.get()
    length = y.get()
    complexity = int(scale4.get())

    
    if username=='':
        messagebox.showerror("Error","Please enter the username.")
        return
    if length==0:
        messagebox.showerror("Error","Please enter password length.")
        return


    # if complexity == 1:
    #     characters = string.ascii_letters
    # elif complexity == 2:
    #     characters = string.ascii_letters + string.digits
    # elif complexity == 3:
    #     characters = string.ascii_letters + string.digits + string.punctuation
    # else:
    #     characters = string.ascii_letters + string.digits + string.punctuation
    
    if complexity >= 1:
        characters = string.ascii_letters
    if complexity >= 2:
        characters += string.digits
    if complexity >= 3:
        characters += string.punctuation

    password = ""
    for i in range(length):
        password += random.choice(characters)
        p_lbl.config(text=f"Generated Password: " + password)


lbl1=Label(root,text='Password Generator',font=('aerial',18),fg='brown')
lbl1.place(x=100,y=0)

lbl2=Label(root,text='User name:',font=('aerial',15))
lbl2.place(x=10,y=70)
x=StringVar()
ent2=Entry(root,textvariable=x)
ent2.place(x=180,y=70,height=30,width=200)

lbl3=Label(root,text='Password length:',font=('aerial',15))
lbl3.place(x=10,y=135)
y=IntVar()
ent3=Entry(root,textvariable=y)
ent3.place(x=180,y=135,height=30,width=200)

lbl4 =Label(root, text="Password Complexity:",font=('aerial',15))
lbl4.place(x=10,y=195)
scale4 =Scale(root, from_=1, to=4, orient=HORIZONTAL)
scale4.place(x=250,y=180)

p_lbl=Label(root,text='',font=('aerial',15))
p_lbl.place(x=10,y=340)



but=Button(root,text='Generate Password',command=generate_password,font=("arial",12),background='brown',foreground='white',activebackground='white',activeforeground='black')
but.place(x=125,y=280,height=40,width=160)

root.mainloop()
```
